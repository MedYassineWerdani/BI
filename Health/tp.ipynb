```python
def group_by_age(age):
    if age < 20 : 
        return "Teen"
    elif 20 <= age <=23:
        return "Young-adult"
    elif age > 23:
        return "Adult"
```

```python
df["Age_group"] = [group_by_age(i) for i in df["Age"]]
```

```python
import sqlite3
con = sqlite3.connect("tutorial.db")
```

```python
cur = con.cursor()
```

```python
cur.execute('''
CREATE TABLE if not exists StudentInfo (
StudentID integer primary key
,Gender Text
,Age Integer"
,Course Text"
,YearOfStudy Text"
,MaritalStatus Text
            ''')
con.commit()
```

```python
cur.execute('''
CREATE TABLE if not exists StudentMentalHealth (
TimeStamp date,
StudentID integer ,
CGPA float ,
Depression Text,
Anxiety Text,
PanicAtack text,
SeekTreatment text,
foreign key (studentID) references StudentInfo(studentID))
            ''')

con.commit()
```

```python
cur.execute(''' 
SELECT COUNT(*) AS NbEtudiants
FROM StudentMentalHealth
WHERE Depression = 'Yes'
AND SeekTreatment = 'Yes';
''' )


result = cur.fetchone()
count = result[0]

print("Nombre d'tudiants :", count)
```

```python
cur.execute(''' 
SELECT * 
FROM StudentInfo as SI
JOIN StudentMentalHealth as SMH
ON SMH.StudentID = SI.StudentID
WHERE Depression = 'Yes'
''' )
result = cur.fetchall()
for row in result: 
    print(row)
```

```python
cur.execute(''' 
SELECT Course , 100.0* SUM(CASE when depression = "Yes" then 1 else 0 END) /Count(*) 
FROM StudentInfo as SI
JOIN StudentMentalHealth as SMH
ON SMH.StudentID = SI.StudentID
GROUP BY Course ;
''' )
result = cur.fetchall()

print("SPECIALTY  :\n\t   PERCENTAGE")
for row in result : 
    print(f"{row[0]} \n\t {row[1]}\n\n") 


```
