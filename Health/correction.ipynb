```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r".\data.csv")
```

```python
df = df.drop_duplicates()
```

```python
df = df[df["Gender"] == 'Female']
```

```python
df.head()
```

```python
#Methode 1
def group_by_age(age):
    if age < 20 : 
        return "Teen"
    elif 20 <= age <=23:
        return "Young-adult"
    elif age > 23:
        return "Adult"
    
df["Student-Health"] = df["Age"].apply(group_by_age)
```

```python
#Methode 2
bins = [15,20,25,float('inf')]
labels = ["15-20","21-25","+25"]
df["Student-Health"] = pd.cut(df["Age"], bins=bins , labels=labels , right=False)
```
